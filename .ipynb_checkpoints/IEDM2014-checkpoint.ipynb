# Plot Capacity Overlap

```python
#==============================================================================
# Generators
#==============================================================================
def MultiGaussianGenerator(N, limits=[4,7], stddev=1):
    '''
    Fit N gaussians the range limits[0]->limits[1]
    equally spaced with stddev
    '''
    Px = np.ones(N) * 1/N
    centers = (np.arange(N) / (N-1)) * (limits[1]-limits[0]) + limits[0]
    Pyx = blahut.FunctionList([lambda x, c=c: stats.norm.pdf(x, loc=c, scale=stddev) for c in centers])
    return Px, Pyx


def plotDensity( Px, Pyx, limits=[-5,15], no_ticks=False, fill=True,
                color_list=['r', 'y', 'g', 'c'], plot_Pyx=False, cmap=False):
    '''
    Plots distributions given Px[array], and Pyx[CapacityTools.FunctionList]
    '''
    ax = gca()
    PxPyx = Pyx * Px    
    
    x = np.linspace(limits[0], limits[1], 1000)
    
    if plot_Pyx:
        Pplot = Pyx
    else:
        Pplot = PxPyx
        
    Pplot = [p for i, p in enumerate(Pplot) if Px[i] != 0]        
        
    for i, p in enumerate(Pplot):
        if cmap:
            color = cmap(i/len(Pplot))
        else:
            color = color_list[i%len(color_list)]
        
        plt.plot(x, p(x), '--', color=color)
#        if i % 2 == 0:
#            plt.plot(x, p(x), '--', color=color)
#        else:
#            plt.plot(x, p(x), ':', color=color)

        if fill:
            plt.fill_between(x, p(x), color=color, alpha=0.3)

    if no_ticks:
        ax.get_xaxis().set_ticks([])
        ax.get_yaxis().set_ticks([])
        
    
    
def plotDensityDiscrete( Px_discrete, Pyx_discrete ):
    '''
    Plot Discrete Pyx Matrices
    '''
    imshow(np.log10(Pyx_discrete), cmap=plt.cm.Blues, interpolation='none')
```

```python
#==============================================================================
#%% Plot #0: 2 Gaussians a)Separate, b)Overlapping
#==============================================================================

JessePlot.rc('ipynb')

subplot(231)
N=2
limits = [5, 7]
scale = 0.22
Px, Pyx = MultiGaussianGenerator(N=N, limits=limits, stddev=scale)
plotDensity(Px, Pyx, limits=[4,8], no_ticks=True, fill=True, color_list=['r', 'b'])

for i, line in enumerate(gca().lines):
    if i % 2 == 0:
        line.set_dashes([4, 2])
    else:
        line.set_dashes([2, 1])
    line.set_linewidth(1)
    
ylabel('P(R|V)')
title('1 Bit')


subplot(232)
N=2
limits = [5, 5.5]
scale = 0.22
Px, Pyx = MultiGaussianGenerator(N=N, limits=limits, stddev=scale)
plotDensity(Px, Pyx, limits=[3.25,7.25], no_ticks=True, fill=True, color_list=['r', 'b'])
xlabel('R')
title('0.6 Bits')

for i, line in enumerate(gca().lines):
    if i % 2 == 0:
        line.set_dashes([4, 2])
    else:
        line.set_dashes([2, 1])
    line.set_linewidth(1)


subplot(233)
N=2
limits = [5.25, 5.25]
scale = 0.22
Px, Pyx = MultiGaussianGenerator(N=N, limits=limits, stddev=scale)
plotDensity(Px, Pyx, limits=[3.25,7.25], no_ticks=True, fill=True, color_list=['r', 'b'])
title('0 Bits')

for i, line in enumerate(gca().lines):
    if i % 2 == 0:
        line.set_dashes([4, 2])
    else:
        line.set_dashes([2, 1])
    line.set_linewidth(1)


tight_layout()
show()
# JessePlot.saveFig('./figs/3gaussians.png')
```

```python
#==============================================================================
# %%load Power Data
#==============================================================================

# Import transistor Data
f = load('./data/transistorFD12.npz')
Vg = f['Vg']
I = f['Isd'].T[-1]
I_spline = interpolate.UnivariateSpline(Vg, I)


rho_Cu = 2e-6 * 1e-2 # ohm-m
eps = 20 * 8.85e-12 # F / m
H = W = S = T = 130e-9 # m
pitch_wl = 1200e-9 # m
pitch_bl = 1400e-9 # m
V_bl = 3 # V
t_pulse = 300e-9 # s


R_wire = lambda L: rho_Cu * L / (H * W) 

C_wire = lambda L: eps * L / 2 * ( 1.15 * W/H + 2.80 * (T/H) ** 0.222 ) + \
         eps * L * 2 * ( 0.03 * (W/H) + 0.83*(T/H) - 0.07 * (T/H) ** 0.222 ) * (H/S) ** 1.34
         
C_transistor = 3e-15 # F / um**2 (assume C_transistor = 1 um**2)


def wire_energy(n_devices, V_wl):
    # Create a square array
    n_side = ceil(sqrt(n_devices))    
    L_wl = pitch_wl * n_side 
    L_bl = pitch_bl * n_side 
            
    # Use full L, since 
    C_wl = C_transistor * n_side + C_wire(L_wl) 
    C_bl = C_wire(L_bl) 

    # Divide L by 2 to average across array
    R_bl = R_wire(L_bl / 2)
    I = I_spline(V_wl)
    
    E_dynamic_wl = C_wl * V_wl**2
    E_dynamic_bl = C_bl * V_bl**2
    E_static_bl = I**2 *R_bl * t_pulse

    return E_dynamic_wl + E_dynamic_bl + E_static_bl
```

```python
#==============================================================================
# %% 3 different sized arrays
#==============================================================================

V_wl= linspace(1.0, 3.0, 100)
JessePlot.rc('ipynb')
for n_devices in [1e3, 1e6, 1e9]:
    semilogy(V_wl, wire_energy(n_devices, V_wl))
xlabel('V$_{WL}$')
ylabel('E (Joules)')
tight_layout()
show()
```
